# Telescope transmission

```python
import numpy as np
import astropy.units as u
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Silver mirrors

```python
from arte.photometry.transmissive_elements_catalogs import EltTransmissiveElementsCatalog

ag_mirror = EltTransmissiveElementsCatalog.ag_mirror_elt_001()

mpl.rcParams['figure.dpi'] = 150
ag_mirror.plot(absorptance=False, transmittance=False, wv_unit=u.um)
plt.xlim(0.5, 2.5)
plt.ylim(0.9, 1.0)
plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Reflectance')
plt.legend('', frameon=False)

wv = ag_mirror.waveset
id_589nm = np.where(np.isclose(np.array(wv), 5890., atol=10))
print(wv[id_589nm])
print(ag_mirror.reflectance(wv[id_589nm]))
```
